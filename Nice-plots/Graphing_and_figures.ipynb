```python
# Notice that lines starting with "#" are ignored

# Next, we need to import some libraries that have some useful functions we can use
import matplotlib
import numpy as np # We can rename a library. Here, we can call numpy "np"
import matplotlib.pyplot as plt
import matplotlib as mpl
plt.style.use('our_plot_style.txt')
%matplotlib inline
```

```python
# How do we plot a graph?

# Start by defining an independent variable
x = np.linspace(0, 3*np.pi, 500)

# And use this function to plot sin(x^2)

fig=plt.figure(figsize=(6, 4)) # size of the figure
plt.plot(x, np.sin(x**2), linewidth = 1, color = "#9c3064")

# We can change the features of the plot, including the title
#plt.title('A simple plot', fontsize=12);

# and the axis labels
plt.xlabel(r'$x$');
plt.ylabel(r'$\sin(x^2)$');
plt.savefig("sample-plot.pdf",format = "pdf",bbox_inches="tight")
```

### Histogram
___________________________________________________________________

#### We can also create a histogram

```python
# Let's take a
#[Gaussian distribution](https://numpy.org/doc/stable/reference/random/generated/numpy.random.normal.html)
# with these parameters:
mean = 0
standard_deviation = 1
num_samples = 5000

# Draw random numbers from this distribution
random_samples = np.random.normal(mean, standard_deviation, num_samples);

# Use matplotlib's hist function to plot a histogram with automatic bins
fig=plt.figure(figsize=(6, 4)) # size of the figure
plt.hist(random_samples, bins=10);

# To save the histogram, do this:
#from google.colab import files
#plt.savefig('HistDemo.pdf',format='pdf',transparent=True)
#files.download('HistDemo.pdf')
```

### Save/load data files (".txt" format)

```python
# Creating 2D array
List = [[1, 2, 3], [4, 5, 6], [7, 8, 9]]
Array = np.array(List)
 
# Displaying the array
print('Array:\n', Array)
 
# Saving the 2D array in a text file
np.savetxt("data-file.txt", Array)
 
# Displaying the contents of the text file
content = np.loadtxt('data-file.txt')
print("\nContent in data-file.txt:\n", content)
```
